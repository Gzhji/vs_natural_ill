```python
from __future__ import print_function
from __future__ import division
from lib2to3.pgen2.token import tok_name
import cv2 as cv
import argparse
import os
import numpy as np
import rawpy
import imageio
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
def loadExopsureSeq(path):
    images = []
    times = []
    with open(os.path.join(path, 'list.txt')) as f:
        content = f.readlines()
        print(content)

        for line in content:
            tokens = line.split()
            rgb = cv.imread(os.path.join(path, tokens[0]))
            images.append(rgb)
            times.append(float(tokens[1]))
            # print(rgb.shape)
    return images, np.asarray(times, dtype=np.float32)
```

```python
def ResponseCurve(path):
    images, times = loadExopsureSeq(path)
    calibrate = cv.createCalibrateDebevec()
    response = calibrate.process(images, times)
    np.savetxt(path + "response_curve.csv", response.reshape((-1,3)), fmt="%s")
    return response
```

```python
theta_pano_directory = 'theta_pano/cropped'
fisheye_directory = '180_fisheye/cropped'

theta_pano_images, theta_pano_times = loadExopsureSeq(theta_pano_directory)
fisheye_images, fisheye_times = loadExopsureSeq(fisheye_directory)

theta_pano_curv = ResponseCurve(theta_pano_directory)
fisheye_curv = ResponseCurve(fisheye_directory)
```

```python
x = np.arange(256, dtype=np.uint8)
y= np.squeeze(theta_pano_curv)

ax = plt.figure(figsize=(5,5))
plt.title('Camera Resoponse Curve - theta_pano', fontsize=10)
plt.xlabel('Measured Pixel', fontsize=10)
plt.ylabel('Calibrated Intensity', fontsize=10)
plt.xlim([0,260])
plt.ylim([0,20])
plt.grid()
plt.plot(x, y[:,0], 'r', label="R")
plt.plot(x, y[:,1], 'g', label="G")
plt.plot(x, y[:,2], 'b', label="B")
plt.legend(loc="upper left", fontsize=10)
```

```python
x = np.arange(256, dtype=np.uint8)
y= np.squeeze(fisheye_curv)

ax = plt.figure(figsize=(5,5))
plt.title('Camera Resoponse Curve - fisheye', fontsize=10)
plt.xlabel('Measured Pixel', fontsize=10)
plt.ylabel('Calibrated Intensity', fontsize=10)
plt.xlim([0,260])
plt.ylim([0,20])
plt.grid()
plt.plot(x, y[:,0], 'r', label="R")
plt.plot(x, y[:,1], 'g', label="G")
plt.plot(x, y[:,2], 'b', label="B")
plt.legend(loc="upper left", fontsize=10)
```
